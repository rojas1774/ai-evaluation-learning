```python
import pandas as pd

data = {
    "response_id": [1, 2, 3, 4, 5],
    "score": [8, 6, 9, 7, 5]
}

df = pd.DataFrame(data)

print("Average score:")
print(df["score"].mean())

print("Highest score:")
print(df["score"].max())

print("Lowest score:")
print(df["score"].min())

print("Responses above 7:")
print(df[df["score"] > 7])
```
